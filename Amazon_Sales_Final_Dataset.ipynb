```python
# import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import Dataset after cleaning in Excel
df=pd.read_csv("/content/Amazon Sales Final.csv")
df
```

```python
# Question 1: What are the dimensions of the dataset?
print("Dataset Dimensions:\n")
df.shape
```

```python
# Question 2: What are the column names?
print("Column Names:\n")
df.columns
```

```python
# Question 3: Are there any missing values in the dataset?
print("Missing Values:\n")
df.isnull().sum()
```

```python
# Question 4: What are the data types of the columns?
print("Column Data Types:\n")
df.dtypes
```

```python
# Question 5: What are the summary statistics of the dataset?
print("Summary Statistics of Dataset:\n")
df.describe()
```

```python
# Question 6: What is the distribution of sales (discounted price vs. actual price)?
plt.figure(figsize=(20,6))
sns.histplot(df['discounted_price'], bins=30, color='Blue', label='Discounted Price')
sns.histplot(df['actual_price'], bins=30, color='red', label='Actual Price')
plt.title('Distribution of Sales')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
# Question 7: What is the average rating and rating count for products?
Average_Rating = df['rating'].mean()
Average_Rating_Count = df['rating_count'].mean()
print("Average Rating:\n", Average_Rating)
print("Average Rating Count\n", Average_Rating_Count)
```

```python
# Question 8: What are the top-selling products by category?
Top_Selling_Products = df.groupby('category')['product_name'].count().nlargest(5)
print("Top 5 Selling Products:\n", Top_Selling_Products)
```

```python
# Question 9: What is the distribution of products across different categories?
plt.figure(figsize=(15,5))
sns.countplot(x='category', data=df)
plt.title('Distribution of Products by Category')
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 10: What is the correlation between ratings, rating count, and discounted price?
correlation_matrix = df[['rating', 'rating_count', 'discounted_price']].corr()
plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Question 11: What is the average discount percentage across all products?
# Since percentage is defined as object, first convert data type in to float
df['discount_percentage'] = df['discount_percentage'].str[:-1].astype('float')
df
```

```python
# Question 11: What is the average discount percentage across all products?
Average_Percentage = df['discount_percentage'].mean()
print("Average Discount Percentage:\n", Average_Percentage)
```

```python
# Question 12: What are the top-rated products with a high number of ratings?
Top_Rated_Products = df[df['rating'] >=4.5].nlargest(5, 'rating_count')[['product_name', 'rating', 'rating_count']]
print("Top Rated Preoducts with high Number of Ratings:\n", Top_Rated_Products)
```

```python
# Question 13: What is the distribution of products by discount percentage?
plt.figure(figsize=(15,5))
sns.histplot(df['discount_percentage'], bins=20, color='Green')
plt.title('Distribution of Discount Percentage')
plt.xlabel('Discount Percentage')
plt.ylabel('Count')
plt.show()
```

```python
# Question 14: What is the average discount percentage by category?
Average_Discount_Percentage = df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)
print("Average Discount Percentage by Category:\n", Average_Discount_Percentage)
```

```python
# Question 15: What is the distribution of ratings for products?
plt.figure(figsize=(8, 6))
sns.histplot(df['rating'], bins=10, color='purple')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Question 16: What is the average rating by category?
Average_Rating_by_Category = df.groupby('category')['rating'].mean().sort_values(ascending=False)
print("Average Rating by Category:\n", Average_Rating_by_Category)
```

```python
# Question 17: What are the most common rating counts for products?
Common_Rating_Counts = df['rating_count'].value_counts().nlargest(5)
print("Average Rating Counts:\n", Common_Rating_Counts)
```

```python
# Question 18: What is the distribution of user ratings?
plt.figure(figsize=(15, 6))
sns.countplot(x='rating', data=df, palette='Blues_r')
plt.title('Distribution of User Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Question 19: What are the top-rated products with the highest discounts?
Top_Rated_Discounted_Products = df[df['rating'] >= 4.5].nlargest(5, 'discount_percentage')[['product_name', 'rating', 'discount_percentage']]
print("Top-rated products with the highest discounts:")
print(Top_Rated_Discounted_Products)
```

```python
# Question 20: What is the average discounted price by category?
avg_discounted_price_by_category = df.groupby('category')['discounted_price'].mean().sort_values(ascending=False)
print("Average Discounted Price by Category:")
print(avg_discounted_price_by_category)
```
